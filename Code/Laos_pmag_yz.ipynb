# Notebook for Laos Basalts

This notebook includes data from insitu basalt flows and exsitu conglomerates interpreted by Sieh et al. 2020 as ejecta that were impacted by the Australasian bolide in southern Laos

### Laos Basalt sample descriptions:

+ no bedding orientations were taken for most sites as they are either near horizontal flows or conglomerates

+ minimal to significant weathering rinds and laterite present on most sampled basalts due to high humidity and monsoon rains
+ diamicton refers to a massive loess/airfall deposit that has been associated with regional fires and associated with bollide due to presence of tektites

+ all ages are Ar/Ar from Sieh et al. 2020 and 

+ site NWFA is 5 block samples from a supposed insitu flow aged 518 ± 6 ka

+ site NWFB is 3 block samples, no sun, from a supposed insitu flow ages 936 30 ka

+ site NWFC is 8 block samples of supposed ejecta boulders in diamicton thought to be between 800-1397 ka

+ site PKQA is 10 core samples of supposed ejecta boulders in diamicton thought to be between 800-1397 ka. These cores are from the same boulder: Core 1,2 | Core 3,4 | Core 5,6 | Core 7,8,9,0

+ site PKSB is 1 block sample from a truly insitu cohernt basalt flow thought to be "shockwave brecciated" aged ~1000 ka

+ site PKSF is 11 core samples of supposed ejecta boulders in diamicton thought to be between 800-1397 ka

+ site PKSG is 8 block samples, no sun, of supposed ejecta boulders in diamicton thought to be between 800-1397 ka

+ site PKSH is 3 core samples, H1 and H2 (aged a little older than 770 ± 8 ka) are in the same horizon 5m above H3 and 12m west. Thought to capture Brunhes-Matuyama reversal at ~780 ka

+ site PKSI is 1 block sample from a questionally insitu corestone aged 553 ± 20 ka

+ site PKSJ is 3 cores samples, no sun, from supposed insitu basalt flow potentially with pahoehoe lobes and comlumnar jointing. aged 601 ± 44 ka

+ site PKSK is 3 core samples, no sun, from  supposed insitu basalt flow from a nearby scoria cone aged 407 ±  18 ka 

+ site WBEA is 2 block samples, no sun, from a massive basalt flow to be used as basalt stratigraphic control. Best outcrop. aged 1397 ka

==== SWILLP EXAMPLES ====

site CS7 has a large pdf. This overprint component can be removed by liquid nitrogen dunk and heating up to ~250 C. Only two specimens (4 and 5 in both a and b series) have a primary component unblocking ca. 500-580 C. In both series these two samples give similar fitted directions. Other samples have largely been completely overprinted such that a direction similar to those of -4 and -5 is hinted in the thermal demag steps but not resolved.

site CS8 was demagnetized in a series only. It has a pdf component that can be removed by up to 475 C, followed by by nice consistent sharply unblocking primary magnetite component from ~530 C to 580 C. There is an additional higher temperature component that is not well resolved.

site CS9 was both demagnetized in a and b series. It has a large pdf overprint that can be removed by LN2 dunk and heating up to ~150-250 C. It looks like the a series without liquid nitrogen dunk is resulted in cleaner data than the b series. This could be due to the b series containing more end chips and thus have overall weaker total magnetization than a series specimens. Another possibility which I think is more likely is that the liquid nitrogen dunk actually resulted in randomization of the magnetite grains that carry the higher temperature component magnetization which in this case caused noise that has high ratios in intensity relative to the signal. This resulted in only three specimens having interpretable mt component in the b series versus 5 in the a series. We use a series data for the mt fits.

site CS10 was both demagnetized in a and b series. It has a pdf component that can be removed largely by LN2 dunk or by heating up to 300-400 C. It appears to have a consistent easterly overprint component that unblocks through magnetite Curie temperature.

site CS11 was both demagnetized in a and b series. It doesn't have consistent site level directions. A few specimens have an easterly overprint component that unblocks through magnetite Curie temperature similar to those in site CS10.

site CS12 was both demagnetized in a and b series. It has a large pdf overprint that can be removed largely by LN2 dunk or by heating up to 300-400 C. LN2 helps removing the PDF overprint component and more specimens yielded interpretable mt components.

site CS13 was both demagnetized in a and b series. It has a large pdf overprint that can be removed by LN2 dunk and heating up to ~150-250 C. It looks like the a series without liquid nitrogen dunk is resulted in cleaner data than the b series. This could be similar to site CS9.

site CS14 was both demagnetized in a and b series. It does not have very consistent directions except for a lt component.

site CS15 was both demagnetized in a and b series. It has a large pdf overprint that can be removed by LN2 dunk and heating up to ~250-400 C. There is not a consistent component that can be fitted through higher temperatures.

```python
# ## 3. Prepare Data
site_lat_dict = dict(zip(Laos_site_locations['Site'], Laos_site_locations['Latitude']))
site_lon_dict = dict(zip(Laos_site_locations['Site'], Laos_site_locations['Longitude']))

def ensure_dash(site_dict):
    return {k if k.endswith('-') else k + '-': v for k, v in site_dict.items()}

site_lat_dict = ensure_dash(site_lat_dict)
site_lon_dict = ensure_dash(site_lon_dict)

Laos_specimen_data['site'] = Laos_specimen_data['sample'].apply(lambda x: '-'.join(x.split('-')[:2]) if '-' in x else x)
Laos_specimen_data['site'] = Laos_specimen_data['site'].str.rstrip('-')
Laos_specimen_data['site'] = Laos_specimen_data['site'].apply(lambda x: x if x.endswith('-') else x + '-')

# Get highest temperature measurement per specimen
highest_temp_measurements = Laos_measurement_data.sort_values('treat_temp').groupby('specimen').last().reset_index()
specimen_directions = highest_temp_measurements[['specimen', 'dir_dec', 'dir_inc']].copy()

# Merge specimen and direction data
merged_data = pd.merge(Laos_specimen_data[['specimen', 'site']], specimen_directions, on='specimen', how='inner')

# ## 4. Plotting Functions

def plot_site_directions(site_code):
    """Plot all specimen directions for a site and the Fisher mean."""
    site_data = merged_data[merged_data['site'] == site_code]
    if len(site_data) == 0:
        print(f"No data available for site {site_code}")
        return
    site_location = Laos_site_locations[Laos_site_locations['Site'] == site_code]
    if len(site_location) == 0:
        print(f"Location information not found for site {site_code}")
        return
    site_lat = site_location.iloc[0]['Latitude']
    site_lon = site_location.iloc[0]['Longitude']
    if len(site_data) < 2:
        print(f"Not enough directional data for Fisher statistics at site {site_code}")
        site_mean = None
    else:
        site_mean = ipmag.fisher_mean(site_data['dir_dec'].tolist(), site_data['dir_inc'].tolist())
        ipmag.print_direction_mean(site_mean)
    plt.figure(figsize=(4, 4))
    ipmag.plot_net(1)
    ipmag.plot_di(site_data['dir_dec'].tolist(), site_data['dir_inc'].tolist(), color='grey', label=f'{site_code} specimen directions')
    if site_mean:
        ipmag.plot_di_mean(site_mean['dec'], site_mean['inc'], site_mean['alpha95'], color='red', label=f'{site_code} site mean')
    ipmag.plot_di(dec=0, inc=ipmag.inc_from_lat(site_lat), marker='*', markersize=60, color='black', label='present dipole field')
    plt.legend(fontsize=11, loc='lower left')
    plt.title(f"Site {site_code} - Core Directional Data")
    plt.savefig(f'./Laos_directions_{site_code}.png', dpi=300)
    plt.show()
    return site_mean

def plot_site_lt_mean(site_prefix):
    specimens_df = pd.read_csv('/Users/nicolasanderson/Library/CloudStorage/OneDrive-CaliforniaInstituteofTechnology/Anderson/CIT/specimens.txt', sep='\t', header=1)
    # normalize for robust matching
    specimens_df['dir_comp_norm'] = specimens_df['dir_comp'].astype(str).str.lower().str.strip()
    specimens_df['specimen_norm'] = specimens_df['specimen'].astype(str).str.strip()
    mask = (
        specimens_df['dir_comp_norm'].str.contains('lt', na=False) &
        (
            specimens_df['specimen_norm'].str.startswith(site_prefix) |
            specimens_df['specimen_norm'].str.startswith(site_prefix + '-')
        )
    )
    site_specimens = specimens_df[mask]
    print(f"[LT] {site_prefix}: matched {len(site_specimens)} rows")
    if len(site_specimens) == 0:
        print(f"No LT measurements found for site {site_prefix}")
        return

    plt.figure(figsize=(4, 4))
    ipmag.plot_net(1)
    ipmag.plot_di(site_specimens['dir_dec'].tolist(), site_specimens['dir_inc'].tolist(),
                  color='grey', alpha=0.5, label='LT measurements')

    if len(site_specimens) > 1:
        mean = ipmag.fisher_mean(site_specimens['dir_dec'].tolist(), site_specimens['dir_inc'].tolist())
        ipmag.plot_di_mean(mean['dec'], mean['inc'], mean['alpha95'], color='green', label='Fisher mean')

        dec = mean.get('dec', float('nan'))
        inc = mean.get('inc', float('nan'))
        n = int(mean.get('n', len(site_specimens)))
        k = mean.get('k', float('nan'))
        alpha95 = mean.get('alpha95', float('nan'))
        print(f"Site {site_prefix} LT Fisher mean -> dec: {dec:.2f}, inc: {inc:.2f}, n: {n}, alpha95: {alpha95:.2f}, k: {k:.2f}")
    else:
        # single direction: report the measured direction
        dec = site_specimens['dir_dec'].iloc[0]
        inc = site_specimens['dir_inc'].iloc[0]
        print(f"Site {site_prefix} LT single direction -> dec: {dec:.2f}, inc: {inc:.2f}, n: 1")

    plt.legend(fontsize=11, loc='lower left')
    plt.title(f'{site_prefix} - LT Measurements & Fisher Mean')
    plt.savefig(f'./CIT_{site_prefix}_lt_mean.png', dpi=300)
    plt.show()

def plot_site_mt_mean(site_prefix):
    specimens_df = pd.read_csv('/Users/nicolasanderson/Library/CloudStorage/OneDrive-CaliforniaInstituteofTechnology/Anderson/CIT/specimens.txt', sep='\t', header=1)
    # normalize for robust matching
    specimens_df['dir_comp_norm'] = specimens_df['dir_comp'].astype(str).str.lower().str.strip()
    specimens_df['specimen_norm'] = specimens_df['specimen'].astype(str).str.strip()
    mask = (
        specimens_df['dir_comp_norm'].str.contains('mt', na=False) &
        (
            specimens_df['specimen_norm'].str.startswith(site_prefix) |
            specimens_df['specimen_norm'].str.startswith(site_prefix + '-')
        )
    )
    site_specimens = specimens_df[mask]
    print(f"[MT] {site_prefix}: matched {len(site_specimens)} rows")
    if len(site_specimens) == 0:
        print(f"No MT measurements found for site {site_prefix}")
        return
    plt.figure(figsize=(4, 4))
    ipmag.plot_net(1)
    ipmag.plot_di(site_specimens['dir_dec'].tolist(), site_specimens['dir_inc'].tolist(), color='grey', alpha=0.5, label='MT measurements')
    if len(site_specimens) > 1:
        mean = ipmag.fisher_mean(site_specimens['dir_dec'].tolist(), site_specimens['dir_inc'].tolist())
        ipmag.plot_di_mean(mean['dec'], mean['inc'], mean['alpha95'], color='purple', label='Fisher mean')

        dec = mean.get('dec', float('nan'))
        inc = mean.get('inc', float('nan'))
        n = int(mean.get('n', len(site_specimens)))
        k = mean.get('k', float('nan'))
        alpha95 = mean.get('alpha95', float('nan'))
        print(f"Site {site_prefix} LT Fisher mean -> dec: {dec:.2f}, inc: {inc:.2f}, n: {n}, alpha95: {alpha95:.2f}, k: {k:.2f}")
    else:
        # single direction: report the measured direction
        dec = site_specimens['dir_dec'].iloc[0]
        inc = site_specimens['dir_inc'].iloc[0]
        print(f"Site {site_prefix} LT single direction -> dec: {dec:.2f}, inc: {inc:.2f}, n: 1")

    plt.legend(fontsize=11, loc='lower left')
    plt.title(f'{site_prefix} - MT Measurements & Fisher Mean')
    plt.savefig(f'./CIT_{site_prefix}_mt_mean.png', dpi=300)
    plt.show()


unique_sites = merged_data['site'].unique()
site_prefixes = ['NWF-A', 'NWF-B', 'NWF-C', 'PKS-B', 'PKS-F', 'PKS-G', 'PKS-H', 'PKS-I', 'PKS-J', 'PKS-K', 'PKQ-A', 'WBE-A']

# Directional plots
for site in unique_sites:
    plot_site_directions(site)

```

```python
def plot_site_pole(site_code):
    """
    Calculate and return the VGP pole data for a site.
    """
    site_data = merged_data[merged_data['site'] == site_code]
    if len(site_data) < 2:
        print(f"Not enough data for site {site_code}")
        return None

    # Get site location
    site_location = Laos_site_locations[Laos_site_locations['Site'] == site_code]
    if len(site_location) == 0:
        print(f"Location info not found for site {site_code}")
        return None
    site_lat = site_location.iloc[0]['Latitude']
    site_lon = site_location.iloc[0]['Longitude']

    # Calculate Fisher mean
    site_mean = ipmag.fisher_mean(site_data['dir_dec'].tolist(), site_data['dir_inc'].tolist())

    # Calculate VGP and uncertainties
    vgp = pmag.dia_vgp(site_mean['dec'], site_mean['inc'], site_mean['alpha95'], site_lat, site_lon)
    vgp_lon, vgp_lat, dp, dm = vgp

    return {
        'site': site_code,
        'vgp_lon': vgp_lon,
        'vgp_lat': vgp_lat,
        'dp': dp,
        'dm': dm,
        'dec': site_mean['dec'],
        'inc': site_mean['inc'],
        'alpha95': site_mean['alpha95'],
        'k': site_mean.get('k', 0),
        'n': site_mean.get('n', 0)
    }
```

```python
def plot_pole_dp_dm(map_axis, plon, plat, slon, slat, dp, dm, pole_label='pole',
                    pole_color='k', pole_edgecolor='k', pole_marker='o',
                    markersize=40, legend=True):
    """
    Plot a paleomagnetic pole with uncertainty ellipse on a Cartopy map.
    """
    # Convert dp, dm from degrees to kilometers (for ellipse size, if needed)
    dp_km = dp * 111.32
    dm_km = dm * 111.32

    # Plot the pole
    map_axis.scatter(plon, plat, marker=pole_marker,
                     color=pole_color, edgecolors=pole_edgecolor, s=markersize,
                     label=pole_label, zorder=101, transform=ccrs.PlateCarree())

    # Plot the uncertainty ellipse (approximate as a circle for simplicity)
    ellipse = plt.Circle((plon, plat), dm, color=pole_color, fill=False,
                         lw=2, alpha=0.7, transform=ccrs.PlateCarree(), zorder=100)
    map_axis.add_patch(ellipse)

    if legend:
        plt.legend(loc='lower left')
```

```python
# ...existing code...
from sklearn.cluster import KMeans
import math

def compute_fisher_stats(df):
    """Return dict with n, dec, inc, k, alpha95 (floats) and count."""
    out = {'n': 0, 'dec': np.nan, 'inc': np.nan, 'k': np.nan, 'alpha95': np.nan, 'count': 0}
    if df is None or len(df) == 0:
        return out
    try:
        mean = ipmag.fisher_mean(df['dir_dec'].tolist(), df['dir_inc'].tolist())
        out['n'] = int(mean.get('n', len(df)))
        out['dec'] = float(mean.get('dec', np.nan))
        out['inc'] = float(mean.get('inc', np.nan))
        out['k'] = float(mean.get('k', np.nan))
        out['alpha95'] = float(mean.get('alpha95', np.nan))
        out['count'] = len(df)
    except Exception:
        out['count'] = len(df)
    return out

for site in unique_sites:
    site_data = merged_data[merged_data['site'] == site].dropna(subset=['dir_dec', 'dir_inc'])
    site_location = Laos_site_locations[Laos_site_locations['Site'] == site]
    if len(site_data) < 2 or len(site_location) == 0:
        continue
    site_lat = site_location.iloc[0]['Latitude']
    site_lon = site_location.iloc[0]['Longitude']
    site_age = site_location.iloc[0]['Age (ka)']

    vgp_df = ipmag.vgp_calc(site_data, 'no', 'Longitude', 'Latitude', 'dir_dec', 'dir_inc')
    coords = vgp_df[['vgp_lon', 'vgp_lat']].values
    n_samples = len(coords)
    if n_samples == 0:
        continue

    max_clusters = min(6, n_samples)
    best_n = 1
    best_score = -math.inf
    eps = 1e-6

    # select cluster count by custom per-cluster Fisher score
    for n in range(1, max_clusters + 1):
        if n == 1:
            labels_try = np.zeros(n_samples, dtype=int)
        else:
            kmeans = KMeans(n_clusters=n, n_init=10, random_state=42)
            labels_try = kmeans.fit_predict(coords)

        cluster_scores = []
        valid = True
        for lab in range(labels_try.max() + 1):
            cluster_vgps = vgp_df[labels_try == lab]
            if len(cluster_vgps) < 2:
                valid = False
                cluster_scores.append(-1e6)
                continue
            stats = compute_fisher_stats(cluster_vgps)
            if np.isnan(stats['k']) or np.isnan(stats['alpha95']):
                valid = False
                cluster_scores.append(-1e6)
                continue
            cluster_score = (stats['k'] + eps) / (stats['alpha95'] + eps)
            cluster_scores.append(cluster_score * (len(cluster_vgps) / n_samples))
        overall_score = sum(cluster_scores)
        if not valid and n != 1:
            overall_score -= 1e3
        if overall_score > best_score:
            best_score = overall_score
            best_n = n

    # final clustering
    if best_n == 1:
        labels = np.zeros(n_samples, dtype=int)
    else:
        kmeans = KMeans(n_clusters=best_n, n_init=10, random_state=42)
        labels = kmeans.fit_predict(coords)

    # helper to compute infos per label
    def cluster_info(labels):
        infos = {}
        for lab in range(labels.max() + 1):
            df = vgp_df[labels == lab]
            stats = compute_fisher_stats(df)
            try:
                vgp = pmag.dia_vgp(stats['dec'], stats['inc'], stats['alpha95'], site_lat, site_lon)
                vgp_lon, vgp_lat, dp, dm = map(lambda x: float(x) if x is not None else np.nan, vgp)
            except Exception:
                vgp_lon = vgp_lat = dp = dm = np.nan
            stats.update({'vgp_lon': vgp_lon, 'vgp_lat': vgp_lat, 'dp': dp, 'dm': dm})
            infos[lab] = stats
        return infos

    infos = cluster_info(labels)

    # merge problematic clusters iteratively
    def find_problematic(infos):
        # problematic if alpha95 == 0, k == inf, or exactly 3 specimens
        return [lab for lab, s in infos.items() if (s.get('alpha95', np.nan) == 0) or np.isinf(s.get('k', np.nan)) or int(s.get('count', 0)) == 3]

    while True:
        infos = cluster_info(labels)
        problematic = find_problematic(infos)
        if not problematic or labels.max() == 0:
            break
        merged_any = False
        for p in problematic:
            if p not in infos:
                continue
            best_partner = None
            best_combined_score = -math.inf
            for q in [lab for lab in infos.keys() if lab != p]:
                merged_df = pd.concat([vgp_df[labels == p], vgp_df[labels == q]], ignore_index=True)
                if len(merged_df) < 2:
                    continue
                merged_stats = compute_fisher_stats(merged_df)
                if np.isnan(merged_stats['k']) or np.isnan(merged_stats['alpha95']):
                    continue
                combined_score = (merged_stats['k'] + eps) / (merged_stats['alpha95'] + eps)
                combined_score *= (len(merged_df) / n_samples)
                if combined_score > best_combined_score:
                    best_combined_score = combined_score
                    best_partner = q
            if best_partner is None:
                sizes = {lab: infos[lab]['count'] for lab in infos}
                partners = [lab for lab in infos if lab != p]
                if partners:
                    best_partner = max(partners, key=lambda x: sizes.get(x, 0))
            if best_partner is None:
                continue
            labels[labels == p] = best_partner
            # reindex labels to contiguous 0..m-1
            uniq = np.unique(labels)
            new_map = {old: new for new, old in enumerate(uniq)}
            labels = np.array([new_map[x] for x in labels])
            merged_any = True
            break
        if not merged_any:
            break

    # final infos after merges
    infos = cluster_info(labels)
    n_clusters = labels.max() + 1
    colors = plt.cm.tab10(np.linspace(0, 1, max(1, n_clusters)))

    # plotting
    plt.figure(figsize=(10, 6))
    ax = plt.axes(projection=ccrs.Mollweide())
    ax.set_global()
    ax.add_feature(cfeature.LAND, facecolor='lightgray', edgecolor='black')
    ax.add_feature(cfeature.OCEAN, facecolor='lightblue')
    ax.add_feature(cfeature.COASTLINE)
    ax.add_feature(cfeature.BORDERS, linestyle=':')
    ax.gridlines()

    for i in range(n_clusters):
        cluster_vgps = vgp_df[labels == i]
        count = len(cluster_vgps)
        if count == 0:
            continue
        ax.scatter(cluster_vgps['vgp_lon'], cluster_vgps['vgp_lat'], color=colors[i], s=40, alpha=0.6,
                   label=f'Mode {i+1} ({count}) Specimens', transform=ccrs.PlateCarree())
        stats = infos.get(i, {})
        if stats.get('count', 0) >= 2:
            plot_pole_dp_dm(ax, stats.get('vgp_lon', np.nan), stats.get('vgp_lat', np.nan),
                            site_lon, site_lat, stats.get('dp', np.nan), stats.get('dm', np.nan),
                            pole_color=colors[i], pole_label=f'Mode {i+1} Mean VGP')
            print(f"Site: {site} | Mode {i+1} | n: {stats.get('count',0)} | VGP Lat: {stats.get('vgp_lat', np.nan):.2f} | VGP Lon: {stats.get('vgp_lon', np.nan):.2f} | α95: {stats.get('alpha95', np.nan):.2f} | k: {stats.get('k', np.nan):.2f}")
        else:
            single = cluster_vgps.iloc[0]
            s_dec = float(single.get('dir_dec', np.nan))
            s_inc = float(single.get('dir_inc', np.nan))
            s_vlon = float(single.get('vgp_lon', np.nan))
            s_vlat = float(single.get('vgp_lat', np.nan))
            ax.scatter(s_vlon, s_vlat, marker='x', color='k', s=80, transform=ccrs.PlateCarree(), zorder=110)
            print(f"Site: {site} | Mode {i+1} | n: {count} (single) | specimen dir dec: {s_dec:.2f}, inc: {s_inc:.2f} | VGP Lat: {s_vlat:.2f}, VGP Lon: {s_vlon:.2f}")

    plt.legend(fontsize=11, bbox_to_anchor=(1, 1))
    plt.title(f'Mollweide VGPs - {site} (KMeans Modes, best_n={n_clusters})')
    plt.savefig(f'./Laos_mollweide_vgps_{site}_kmeans_modes.png', dpi=300)
    plt.show()
# ...existing code...
```

```python

# 2. Merge with site locations for age and depth
specimens_mt = specimens_mt.merge(
    Laos_site_locations[['Site', 'Age (ka)', 'Depth']],
    left_on='site', right_on='Site', how='left'
)
specimens_mt

```

```python
# 3. Aggregate by site (not sample)
mt_summary = specimens_mt.groupby('site').agg(
    mean_inclination=('dir_inc', 'mean'),
    std_inclination=('dir_inc', 'std'),
    n_samples=('dir_inc', 'count'),
    age_ka=('Age (ka)', 'first'),
    depth=('Depth', 'first')
).reset_index()
mt_summary['age_ka'] = pd.to_numeric(mt_summary['age_ka'], errors='coerce')
mt_summary
```

```python
# 4. Add polarity column (all 'mt' here, or assign by sign if needed)
mt_summary['polarity'] = np.where(mt_summary['mean_inclination'] >= 0, 'Normal', 'Reversed')
mt_summary
```

```python

def plot_mt_inclination_vs_time_depth(mt_df):
    # Remove rows with NaN or non-finite mean_inclination or depth
    mt_df = mt_df[np.isfinite(mt_df['mean_inclination']) & np.isfinite(mt_df['depth'])]
    print(f"Plotting {len(mt_df)} rows")  # <--- Add this line
    if mt_df.empty:
        print("No data to plot after filtering.")
        return
    fig, ax = plt.subplots(figsize=(12, 10))
    mt_df_sorted = mt_df.sort_values('depth')
    ax.scatter(mt_df_sorted['mean_inclination'], mt_df_sorted['depth'], 
               c='grey', alpha=0.2, s=30, label='All Sites (mt)', zorder=1)
    normal = mt_df_sorted[mt_df_sorted['polarity'] == 'Normal']
    reversed_ = mt_df_sorted[mt_df_sorted['polarity'] == 'Reversed']
    if len(normal) > 0:
        ax.errorbar(normal['mean_inclination'], normal['depth'],
                    xerr=normal['std_inclination'], fmt='o', color='red',
                    label='Normal Polarity', capsize=5, zorder=2)
    if len(reversed_) > 0:
        ax.errorbar(reversed_['mean_inclination'], reversed_['depth'],
                    xerr=reversed_['std_inclination'], fmt='s', color='blue',
                    label='Reversed Polarity', capsize=5, zorder=2)
    for _, row in mt_df.iterrows():
        x_offset = 12 if row['mean_inclination'] >= 0 else -12
        ha = 'left' if row['mean_inclination'] >= 0 else 'right'
        # Safely format age
        if pd.notnull(row['age_ka']) and isinstance(row['age_ka'], (int, float, np.integer, np.floating)):
            age_str = f"{row['age_ka']:.0f} ka"
        else:
            age_str = "Age N/A"
        ax.annotate(f"{row['site']}\n{row['polarity']}\n{age_str}\nn={row['n_samples']}",
                    (row['mean_inclination'], row['depth']),
                    xytext=(x_offset, 0), textcoords='offset points',
                    fontsize=9, fontweight='bold', ha=ha, va='center',
                    bbox=dict(boxstyle='round,pad=0.4', 
                            facecolor='lightcoral' if row['polarity']=='Normal' else 'lightblue', 
                            edgecolor='black', alpha=0.8),
                    zorder=3)
    ax.axvline(x=0, color='black', linestyle='-', alpha=0.8, linewidth=2, zorder=0)
    ax.axvspan(-90, 0, alpha=0.1, color='blue')
    ax.axvspan(0, 90, alpha=0.1, color='red')
    ax.set_xlabel('Inclination (degrees)', fontsize=14, fontweight='bold')
    ax.set_ylabel('Relative Depth (youngest at top)', fontsize=14, fontweight='bold')
    ax.set_title('MT Geographic Inclinations vs Time vs Relative Depth', fontsize=16, fontweight='bold', pad=20)
    ax.set_xlim(-95, 95)
    ax.invert_yaxis()
    ax.grid(True, alpha=0.3, zorder=1)
    ax.legend()
    ax2 = ax.twinx()
    yticklabels = [f"{age:.0f} ka" if pd.notnull(age) else "N/A" for age in mt_df_sorted['age_ka']]
    ax2.set_yticks(mt_df_sorted['depth'])
    ax2.set_yticklabels(yticklabels, fontsize=11, fontweight='bold')
    ax2.set_ylabel('Age (ka)', rotation=270, labelpad=20, fontsize=14, fontweight='bold')
    ax2.invert_yaxis()
    plt.tight_layout()
    plt.show()

```
